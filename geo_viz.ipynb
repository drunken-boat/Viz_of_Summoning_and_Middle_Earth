```python
import folium
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
import geopandas
import contextily as cx

poi = pd.read_excel('final_geo_data/csv/obvious_poi.xlsx')

gdf = gpd.GeoDataFrame(
    poi, geometry=gpd.points_from_xy(poi.y,poi.x))

gdf.head()

gdf = gdf.set_crs(4326, allow_override=True)

ax = gdf.plot(figsize=(20, 10))
cx.add_basemap(ax, crs=gdf.crs)
cx.add_basemap(ax, crs=gdf.crs, source=cx.providers.Stamen.Watercolor)
cx.add_basemap(ax, source=cx.providers.Stamen.TonerLabels)

plt.savefig('geo_poi.png',dpi=500)
```
