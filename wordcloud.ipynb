```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn
from wordcloud import WordCloud, STOPWORDS

# font_path = 'fonts/aniron/anirm___.ttf'
font_path = 'fonts/morris-roman/MorrisRomanAlternate-Black.ttf'
text = open('lyrics_for_wordcloud.txt').read()

stopwords = set(STOPWORDS)
additional_stopwords = ['Sample:', '1.', '2.', 'Chours', 'Verse', 'Choir:', 'Refrain:']
for i in additional_stopwords:
    stopwords.add(i)

wc = WordCloud(background_color=None, max_words=70, min_font_size=20, max_font_size=195,mode='RGBA',
               stopwords=stopwords, width=800, height=400, font_path=font_path, colormap='Dark2')
wc.generate(text)

# show
# plt.figure(figsize=[20,10])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.style.use('seaborn-colorblind')
# plt.show()
# plt.plot()

plt.savefig('word_cloud_test8.png', dpi=600, bbox_inches = 'tight')
```
